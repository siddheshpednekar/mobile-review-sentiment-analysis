```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('output.csv')
df
```

```python
stars = df['ratings&reviews'].str.split('‚òÖ', expand=True)
stars[0].value_counts()
```

```python
counts = df.groupby(by=["sentiment"]).size().reset_index(name="counts")
counts
```

```python
import plotly.express as px
%matplotlib inline
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_dark"
```

```python
import plotly.express as px
fig = px.bar(data_frame=counts, x="sentiment", y="counts", color="sentiment", text='counts')
fig.update_traces(textfont_size=12, textangle=0, textposition="outside")
```
